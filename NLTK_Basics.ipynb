## Tagging Parts of Speech

```python
words = word_tokenize(example_string)
```

```python
nltk.pos_tag(words)
```

```python
nltk.help.upenn_tagset()
```

```python
df = pd.DataFrame(nltk.pos_tag(words), columns=['word', 'tag'])
```

```python
tag_counts = df.groupby(by='tag').count()
tag_counts
```

```python
tag_counts = tag_counts.sort_values(by='word', ascending=False)
tag_counts
```

```python
plt.bar(x=tag_counts.index.to_list(), height=tag_counts['word'])
plt.xticks(rotation = 90)
plt.xlabel('Category')
plt.title('Word Counts')
plt.show()
```

## Using Named Entity Recognition (NER)

```python
def extract_ne(quote):
    words = word_tokenize(quote, language='English')
    tags = nltk.pos_tag(words)
    tree = nltk.ne_chunk(tags, binary=True)
    return set(
        " ".join(i[0] for i in t)
        for t in tree
            if hasattr(t, "label") and t.label() == "NE"
    )
```

```python
extract_ne(example_string)
```

## Using Concrdance

```python
words = word_tokenize(example_string)
text = nltk.Text(words)
text.concordance("Python")
```

```python
text.concordance("language")
```

## Making a Dispersion Plot

## Making a Frequency Distribution

```python
frequency_distribution = FreqDist(words)
frequency_distribution
```

```python
stop_words = stop_words.union([',', '.', '’', '!'])
```

```python
meaningful_words = lambda text: [word for word in text if word not in stop_words]
```

```python
frequency_distribution = FreqDist(meaningful_words(words))
frequency_distribution
```

```python
frequency_distribution.plot(20, cumulative=False)
```

```python
d = sorted(frequency_distribution.items(), key=lambda item: item[1], reverse=True)
```

```python
idx = [item[0] for item in d][:20]
```

```python
values = [item[1] for item in d][:20]
```

```python
plt.bar(idx, values)
plt.xticks(rotation=90)
plt.title('Word Frequency')
plt.show()
```
