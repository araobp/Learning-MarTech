```python
import fitz
import requests
import re

# 総務省　第7節 ICTサービス及びコンテンツ・ アプリケーションサービス市場の動向
resp = requests.get('https://www.soumu.go.jp/johotsusintokei/whitepaper/ja/r05/pdf/n4700000.pdf')
doc = fitz.open(stream=resp.content)
```

```python
page_num = 1
texts = []
for page in doc:
    output = page.get_text("text")
    text = output.replace('\n', '')
    texts.append(text)
texts
```

```python
for page in doc:
    rects = page.search_for("メタバース")
    page.add_highlight_annot(rects)
```

```python
doc.save('tmp/highlighted.pdf')
```
