```python
from pptx import Presentation

# 私が作成した資料
filename='../material_samples/MyPhotosアプリ概要設計書.pptx'
```

```python
p = Presentation(filename)
```

## テキスト抽出

```python
page = 1
for slide in p.slides:
    print('--- Page {}'.format(page))
    page += 1
    for shape in slide.shapes:
        if hasattr(shape, "text"):
            print(shape.text)
```
