```python
with open('positive_negative.txt', 'r') as f:
    imp = f.read()
print(imp)
```

```python
import spacy
import ginza

ginza.set_split_mode(nlp, 'B')

nlp = spacy.load('ja_ginza')
doc = nlp(imp)
```

```python
for sent in doc.sents:
    for token in sent:
        print(token)
```

```python
for sent in doc.sents:
    for token in sent:
        print(token.i, token.orth_, token.lemma_, token.pos_, 
              token.tag_, token.dep_, token.head.i)
```

```python
for ent in doc.ents:
    print(ent)
```
